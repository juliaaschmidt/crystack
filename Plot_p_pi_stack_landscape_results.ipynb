```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# df = pd.read_csv("./degree_pi_stack/results_12helicene.csv", header=None, names=['cryst_no', 'pi_stacked_rings', 'rings', 'pipi'])
# d = pd.read_csv("12helicene_dmacrys.txt", sep=" ", header=None, names=['cryst_no', 'sg', 'dens', 'E'])
# d
name = "12helicene"
df = pd.read_csv(f"./degree_pi_stack/results_{name}.csv", header=None, names=['cryst_no', 'pi_stacked_rings', 'rings', 'pipi'])
d = pd.read_csv(f"{name}_dmacrys.txt", sep=" ", header=None, names=['cryst_no', 'sg', 'dens', 'E'])
d
```

```python
c = pd.merge(df, d, on="cryst_no")
```

```python
plt.scatter(c.dens, c.pipi*100, c="dodgerblue") #c=c.pipi)
plt.ylabel("% rings with Ï€-Ï€ stacking", fontsize=14) #"Density (g cm$^{-3}$)"
plt.xlabel("Density (g cm$^{-3}$)", fontsize=14)
#plt.xlabel("Lattice Energy (kJ mol$^{-1}$)", fontsize=14)
# cbar = plt.colorbar()#"% rings involved in $\pi$-$\pi$ stacking")
# cbar.set_label('% rings with Ï€-Ï€ stacking', rotation=90, fontsize=12)
plt.savefig(f"{name}_dens.png", dpi=100)
```

```python
plt.scatter(c.E, c.pipi*100, c="dodgerblue") #c=c.pipi)
plt.ylabel("% rings with Ï€-Ï€ stacking", fontsize=14) #"Density (g cm$^{-3}$)"
#plt.xlabel("Density (g cm$^{-3}$)", fontsize=14)
plt.xlabel("Lattice Energy (kJ mol$^{-1}$)", fontsize=14)
# cbar = plt.colorbar()#"% rings involved in $\pi$-$\pi$ stacking")
# cbar.set_label('% rings with Ï€-Ï€ stacking', rotation=90, fontsize=12)
plt.savefig(f"{name}_ener.png", dpi=100)
```

```python
plt.hist(c.pipi*100, bins=10, density=True, color="dodgerblue")#, c=c.pipi)
plt.xlabel("% rings with Ï€-Ï€ stacking", fontsize=14) #"Density (g cm$^{-3}$)"
#plt.vline(np.mean(c.pipi*100))
#plt.ylabel("Lattice Energy (kJ mol$^{-1}$)", fontsize=14)
plt.ylabel("Frequency", fontsize=14)
plt.title(f"{name}")
plt.savefig(f"{name}_histogram_pi_stacking.png",dpi=100)
```
